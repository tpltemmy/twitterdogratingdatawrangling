## Cleaning Data
In this section, clean **all** of the issues you documented while assessing. 

**Note:** Make a copy of the original data before cleaning. Cleaning includes merging individual pieces of data according to the rules of [tidy data](https://cran.r-project.org/web/packages/tidyr/vignettes/tidy-data.html). The result should be a high-quality and tidy master pandas DataFrame (or DataFrames, if appropriate).

```python
# Make copies of original pieces of data
df_clean = df.copy()
df_tweetclean = df_tweet.copy()
df_imageclean = df_image.copy()
```

### Issue #1: There are retweets in the actual data

#### Define: Drop the values that are not null in the retweeted_status_id

#### Code

```python
df_clean = df_clean[~df_clean['retweeted_status_id'].notnull()]
print(df_clean)
```

#### Test

```python
df_clean.info()
```

### Issue #2: Wrong denominator values in some rows

#### Define: All denominator value will be replace with the standard value 10

###### Code 

```python
df_clean['rating_denominator'] = df['rating_denominator'].apply(lambda x: 10)
```

#### Test

```python
df_clean.rating_denominator.value_counts()
```

```python
df_clean.head(5)
```

### Issue #3: Missing and incorrect names

#### Define: Gather the wrong names and replace with general dog name

##### Code 

```python
df_clean['name'] = df_clean['name'].replace(['the','a','an','None','quite','such'],'Dog')
```

#### Test

```python
df_clean.name.value_counts()
```

### Issue 4: Dog Stages are in Strings instead of boolean 

#### Define: Convert all the stages column from strings to boolean 

##### Code 

```python
df_clean['doggo'] = df_clean['doggo'].map({'doggo': True, 'None': False})
df_clean['floofer'] = df_clean['floofer'].map({'floofer': True, 'None': False})
df_clean['pupper'] = df_clean['pupper'].map({'pupper': True, 'None': False})
df_clean['puppo'] = df_clean['puppo'].map({'puppo': True, 'None': False})
```

##### Test 

```python
df_clean.info()
```

### Issue 5: The dog stages variables are in different columns  

#### Define: Aggregate the four columns into one column named Stages 

##### Code 

```python
dog_stages = df_clean[['doggo','floofer','pupper','puppo']]

def dog_stages(row):
   if row['doggo'] == True:
      return 'doggo'
   if row['floofer'] == True:
      return 'floofer'
   if row['pupper'] == True:
      return 'pupper'
   if row['puppo'] == True:
      return 'puppo'
```

```python
df_clean['Stages'] = df_clean.apply(dog_stages, axis=1)
```

##### Test 

```python
df_clean.head(5)
```

```python
df_clean.Stages.value_counts()
```

```python
df_clean = df_clean.drop(labels=['retweeted_status_id', 'retweeted_status_user_id', 'retweeted_status_timestamp',
                               'pupper', 'doggo', 'puppo', 'floofer'], axis = 1)

```
